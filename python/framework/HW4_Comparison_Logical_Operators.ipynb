<h2 style='direction:rtl;text-align:right;font-family:samim;'><b>🔸 تمرین ۱</b></h2>

<p style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>
خروجی گیت XOR برای سه ورودی به صورت زیر محاسبه می‌شود.
<p style='direction:rtl;text-align:center;font-size:18px;font-family:samim;'>
    $$Q = A\overline{BC} + \overline{A} B \overline{C} + \overline{AB}C + ABC $$
<p style='direction:rtl;text-align:right;font-size:18px;font-family:samim;line-height:1.5;'>    
    این گیت را با عملگرهای منطقی پیاده‌سازی کنید و خروجی را با جدول درستی گیت  xor که در زیر آمده‌است چک کنید:
</p>

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>راهنمایی</h4>

<p style='direction:rtl;text-align:right;font-size:16px;font-family:samim;'>
متغیرهای بولین را در ابتدا با مقداری دلخواه مقداردهی کنید. سپس هر جمله را جداگانه پیاده سازی کنید و در نهایت با هم جمع کنید. هر جا ضرب بود با گیت and و هر جمع بود با گیت or پیاده‌سازی کنید.
</p>

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>جواب</h4>

```python
#تعریف متغیرها
a = True * 0
b = True * 1
c = True * 0

#پیاده سازی جمله اول
Q1 = a and not(b) and not(c)
print(Q1)

#پیاده سازی جمله دوم
Q2 = not (a) and b and not(c)
print(Q2)

#پیاده سازی جمله سوم
Q3 = not(a) and not(b) and c
print(Q3)

#پیاده سازی جمله آخر
Q4 = a and b and c
print(Q4)

#کل جمله
Q = Q1 or Q2 or Q3 or Q4
print(Q)
```

<h2 style='direction:rtl;text-align:right;font-family:samim;'><b>🔸 تمرین ۲</b></h2>

<p style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>
در جدول زیر خروجی هر دستور به ازای گزاره  گفته شده چه خواهد بود:
</p>

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>راهنمایی</h4>

<p style='direction:rtl;text-align:right;font-size:16px;font-family:samim;'>
تعاریف any و all را یکبار مطالعه کنید.
سپس لیستی با مقادیر خواسته‌شده در جدول تعریف کنید و any  و  all  را برای آن‌ها اجرا کنید.
</p>

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>جواب</h4>

```python
#True تعریف لیستی با مقادیر همه 
l1 = [True, True , True]

q1 = any(l1)
q2 = all(l1)
print(q1 , q2)

#False تعریف لیستی با مقادیر همه 
l2 = [False, False , False]
q1 = any(l2)
q2 = all(l2)
print(q1 , q2)

# False و بقیه True تعریف لیست با یکی 
l3 = [True, False , False]
q1 = any(l3)
q2 = all(l3)
print(q1 , q2)

# True و بقیه False تعریف لیست با یکی 
l4 = [False, True , True]
q1 = any(l4)
q2 = all(l4)
print(q1 , q2)

#تعریف لیست خالی
l5 = []
q1 = any(l5)
q2 = all(l5)
print(q1 , q2)
```

<h2 style='direction:rtl;text-align:right;font-family:samim;'><b>🔸 تمرین ۷</b></h2>

<p style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>
    با توجه به اینکه خروجی مدار زیر برابر با عبارت H است، آن را با عملگرهای منطقی پیاده‌سازی کنید و جدول درستی را رسم کنید. 
<p style='direction:rtl;text-align:center;font-size:18px;font-family:samim;'>    
$$H = A' + B.C' + A'.B'$$  
</p>

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>راهنمایی</h4>

<p style='direction:rtl;text-align:right;font-size:16px;font-family:samim;'>
ابتدا سه متغیر از جنس صفر یا یک تعریف کنید. سپس رابطه گفته شده را پیاده‌سازی کنید. و در نهایت برای حالات مختلف از ورودی ، خروجی را محاسبه کنید.
</p>

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>جواب</h4>

```python
#تمامی حالاتی که ممکن است برای ورودی ها داشته باشیم را تعریف می کنیم
a = [0, 0, 0, 0, 1, 1, 1, 1]
b = [0, 0, 1, 1, 0, 0, 1, 1]
c = [0, 1, 0, 1, 0, 1, 0, 1]

H_val = []

#هربار برای یکی از حالات ورودی خروجی را محاسبه می کنیم

H_val.append((not a[0]) or (b[0] and not c[0]) or (not a[0] and not b[0]))

H_val.append((not a[1]) or (b[1] and not c[1]) or (not a[1] and not b[1]))

H_val.append((not a[2]) or (b[2] and not c[2]) or (not a[2] and not b[2]))

H_val.append((not a[3]) or (b[3] and not c[3]) or (not a[3] and not b[3]))

H_val.append((not a[4]) or (b[4] and not c[4]) or (not a[4] and not b[4]))

H_val.append((not a[5]) or (b[5] and not c[5]) or (not a[5] and not b[5]))

H_val.append((not a[6]) or (b[6] and not c[6]) or (not a[6] and not b[6]))

H_val.append((not a[7]) or (b[7] and not c[7]) or (not a[7] and not b[7]))

print(H_val)
```

متغیری از جنس رشته برای متن داده شده تعریف کنید. سپس آنرا با دستور .split توکنایز و سپس تبدیل به لیست کنید.    
لیست خرید مشتری و وزن آن‌ها را به عنوان متغیری از جنس لیست تعریف کنید.

<h4 style='direction:rtl;text-align:right;font-size:18px;font-family:samim;'>جواب</h4>

```python
#تعریف لیست میوه ها
fruits = "apple apricot avocado banana blackberry blackcurrant blueberry cherry coconut fig grape lime lychee mandarian mango melon nectarine orange papaya passion peach pear pineapple plum pomegranate raspberry strawberry watermelon"
fruits_list = list(fruits.split())
print(len(fruits_list))

#لیست قیمت میوه ها
cost_0 = [21, 38, 41, 41, 41, 42, 36, 47, 14, 24, 11, 18, 38, 46, 46, 35, 34, 10, 37, 15, 40, 19, 10, 15, 18, 12, 40, 50]
print(len(cost_0))

#تعریف لیست خرید مشتری
customer_list = ['blackberry','cucumber', 'avocado', 'blackcurrant', 'mango' ,'strawberry', 'plum', 'coconut', 'pineapple', 'papaya']
print(len(customer_list))

#لیست قیمت خرید
weight = [1, 3.5, 5, 3, 2.5, 5, 1.5, 4.5, 5, 3]
print(len(weight))

#محاسبه فاکتور خرید
total_cost = (cost_0[fruits_list.index(customer_list[0])]*weight[0] +
             cost_0[fruits_list.index(customer_list[2])]*weight[2] +
             cost_0[fruits_list.index(customer_list[3])]*weight[3] +
             cost_0[fruits_list.index(customer_list[4])]*weight[4] +
             cost_0[fruits_list.index(customer_list[5])]*weight[5] +
             cost_0[fruits_list.index(customer_list[6])]*weight[6] +
             cost_0[fruits_list.index(customer_list[7])]*weight[7] +
             cost_0[fruits_list.index(customer_list[8])]*weight[8] +
             cost_0[fruits_list.index(customer_list[9])]*weight[9])

print(total_cost)
```
