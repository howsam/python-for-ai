## **‚úçÔ∏è Example: Sigmoid**

$$y=\sigma(x)=\frac{1}{1+\exp(-ax)}$$

```python
def sigmoid(x, a):
    y = 1 / (1 + np.exp(-a*x))
    return y
```

```python
x = np.linspace(-10, 10, 1000)
y = sigmoid(x, a=5)

plt.plot(x, y)
```

```python
x = np.linspace(-10, 10, 1000)

plt.plot(x, sigmoid(x, a=0.1))
plt.plot(x, sigmoid(x, a=1))
plt.plot(x, sigmoid(x, a=2))
plt.plot(x, sigmoid(x, a=5))
plt.plot(x, sigmoid(x, a=10))
```

## üî∂ **Section 3: Title**

<img src="../python-notebook/images/title.png" alt="string indexing" width=500 align="center" />

### `plt.title()`

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.title('sin(x)', loc='center', fontsize=15, color='red', fontfamily='serif');
```

```python
x = np.linspace(0, 10, 100)
alpha = 2
plt.plot(x, np.sin(x+alpha))

fontdict={'family':'serif', 'size':15, 'color':'r'}
plt.title(f'function: $sin(x_i + {alpha})$', loc='center', fontdict=fontdict);
```

### `plt.xlabel()` & `plt.ylabel()`

```python
x = np.linspace(0, 10, 100)
alpha = 2
plt.plot(x, np.sin(x+alpha))

fontdict={'family':'serif', 'size':15, 'color':'r'}
plt.title(f'function: $sin(x_i + {alpha})$', loc='center', fontdict=fontdict)

plt.xlabel('$x$', fontdict=fontdict, loc='right')
plt.ylabel('$y$', fontdict=fontdict, loc='top');
```

### `plt.legend()`

```python
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin')
plt.plot(x, np.cos(x), label='cos')

plt.legend(loc='center')
```

## **‚úçÔ∏è Example: Plot gaussian with scatter**

<img src="../python-notebook/images/scatter-example.png" alt="string indexing" width=400 align="center" />

```python
xyb = np.random.randn(1000, 2)
plt.scatter(xyb[:, 0], xyb[:, 1], c='b', edgecolor='w', label='$N(\mu=0, \sigma=1)$')

xyr = 0.5*np.random.randn(1000, 2) + 2
plt.scatter(xyr[:, 0], xyr[:, 1], c='r', edgecolor='w', label='$N(\mu=2, \sigma=0.5)$')

plt.title('Gaussian')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(alpha=0.2)
plt.grid(which='minor', linestyle=':', alpha=0.2)
plt.minorticks_on()
plt.legend();
```

### How to plot a bar graph

```python
x1 = [78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90, 91, 92, 93, 94, 95, 96]
y1 = [7,  12, 14, 16, 15, 13, 7,  8,  3,  0,  2,  11, 7,  11, 11, 12, 9,  12, 23]

plt.bar(x1, y1)
```

### More on bars!

#### You can still have facecolor, edgecolor, etc.

```python
plt.bar(x1, y1, width=1.5, facecolor='r', edgecolor='b', linewidth=1.2, alpha=0.5)
```

#### Plot multiple bar charts

```python
x1 = np.array([78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90, 91, 92, 93, 94, 95, 96])

y1 = np.array([7,  12, 14, 16, 15, 13, 7,  8,  3,  0,  2,  11, 7,  11, 11, 12, 9,  12, 23])
y2 = np.array([9,  10, 18, 5, 3, 4, 0,  2,  3,  4,  1,  6, 8,  9, 10, 5, 7,  11, 6])

plt.bar(x1, y1)
plt.bar(x1, y2, color='r', alpha=0.5)
```

```python
x1 = np.array([78, 79, 80, 81, 82, 83, 84, 85, 86, 87, 88, 89, 90, 91, 92, 93, 94, 95, 96])

y1 = np.array([7,  12, 14, 16, 15, 13, 7,  8,  3,  0,  2,  11, 7,  11, 11, 12, 9,  12, 23])
y2 = np.array([9,  10, 18, 5, 3, 4, 0,  2,  3,  4,  1,  6, 8,  9, 10, 5, 7,  11, 6])

w = 0.4
plt.bar(x1, y1, width=w)
plt.bar(x1-w, y2, width=w)
plt.xticks(ticks=np.arange(78, 97)-w/2, labels=np.arange(78, 97));
```
